#### Determining whether 0 passing yards is missing data by grabbing stats from ESPN.

```python
import requests
from bs4 import BeautifulSoup


def get_team_stats_espn(game_id) :
    url = "https://www.espn.com/college-football/matchup/_/gameId/" + str(game_id)
    head = headers={
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
                       AppleWebKit/537.36 (KHTML, like Gecko) \
                       Chrome/116.0.0.0 Safari/537.36"
    }

    req = requests.get(url, headers=head)

    if(req.status_code != 200) :
        return None

    stat_dict = dict()
    soup = BeautifulSoup(req.text, "html.parser")
    
    teams = [e.text for e in soup.find_all("h2", 
                        {"class" : "ScoreCell__TeamName"})]
    stat_dict["teams"] = teams
    
    table = soup.find("section", attrs={"class" : "Card TeamStatsTable"})
    for entry in table.find_all("tr") :
        name = entry.find("td")
        if(name == None) :
            continue
        stat_dict[name.text] = []
        for el in entry.find_all("td")[1:] :
            stat_dict[name.text].append(el.text)
            
    return stat_dict

def get_stats_for_games_ids_espn(game_ids) :
    stats_df = pd.DataFrame()
    for game_id in game_ids :
        df = pd.DataFrame(get_team_stats_espn(game_id))
        for t in range(1,3) :
            stats_df.loc[game_id, "team"+str(t)] = df.loc[t-1, "teams"]
            stats_df.loc[game_id, "team"+str(t)+"_firstDowns"] = int(df.loc[t-1, "1st Downs"])
            stats_df.loc[game_id, "team"+str(t)+"_thirdDownEff"] = df.loc[t-1, "3rd down efficiency"]
            stats_df.loc[game_id, "team"+str(t)+"_fourthDownEff"] = df.loc[t-1, "4th down efficiency"]
            stats_df.loc[game_id, "team"+str(t)+"_netPassingYards"] = int(df.loc[t-1, "Passing"])
            stats_df.loc[game_id, "team"+str(t)+"_yardsPerPass"] = float(df.loc[t-1, "Yards per pass"])
            stats_df.loc[game_id, "team"+str(t)+"_interceptions"] = int(df.loc[t-1, "Interceptions thrown"])
            stats_df.loc[game_id, "team"+str(t)+"_completionAttempts"] = df.loc[t-1, "Comp-Att"]
            stats_df.loc[game_id, "team"+str(t)+"_fumblesLost"] = int(df.loc[t-1, "Fumbles lost"])
            stats_df.loc[game_id, "team"+str(t)+"_turnovers"] = int(df.loc[t-1, "Turnovers"])
            stats_df.loc[game_id, "team"+str(t)+"_yardsPerRushAttempt"] = float(df.loc[t-1, "Yards per rush"])
            stats_df.loc[game_id, "team"+str(t)+"_rushingAttempts"] = int(df.loc[t-1, "Rushing Attempts"])
            stats_df.loc[game_id, "team"+str(t)+"_rushingYards"] = int(df.loc[t-1, "Rushing"])
    return stats_df
```

#### Add average performance of prior games.

```python
def rolling_mean_stat(col, team_name) :
    team1_select = (merged_df.team1 == team_name)
    team2_select = (merged_df.team2 == team_name)
    team_df = merged_df[team1_select | team2_select]
    team1_select = (team_df.team1 == team_name)
    team2_select = (team_df.team2 == team_name)
    
    rolling_df = pd.DataFrame(index = team_df.index, columns=["def_"+col, "off_"+col])
    
    rolling_df["def_"+col] \
        = ((team1_select)*team_df["team2_"+col] + (team2_select)*team_df["team1_"+col]) \
          .rolling(4, min_periods=1).mean()
    rolling_df["off_"+col] \
        = ((team1_select)*team_df["team1_"+col] + (team2_select)*team_df["team2_"+col]) \
          .rolling(4, min_periods=1).mean()
        
    return rolling_df.shift(1)
```

# Model Selection

```python
# Baseline model : predict the team with the higher pregame_elo will win.
# The predicted probability that team1 will win is 1 if team1 has a higher pregame_elo.

class HigherELOWins :
    def __init__(self) :
        return
    def fit(self, X, y) :
        return self
    def predict(self, X) :
        return (X.team1_pregame_elo > X.team2_pregame_elo)
    def score(self, X, y) :
        return (self.predict(X) == y).sum()/y.shape[0]
    def predict_proba(self, X) :
        res = np.array(self.predict(X).apply(int))
        return np.stack((1-res, res), axis=1)
    def get_params(self, *args, **kwargs) :
        return {}
```

```python
X_columns = ['team1_home_away', 'team2_home_away', 'team1_pregame_elo',
       'team2_pregame_elo', 'team1_off_interceptions',
       'team2_off_interceptions', 'team1_def_interceptions',
       'team2_def_interceptions', 'team1_off_turnovers', 'team2_off_turnovers',
       'team1_def_turnovers', 'team2_def_turnovers',
       'team1_off_yardsPerRushAttempt', 'team2_off_yardsPerRushAttempt',
       'team1_def_yardsPerRushAttempt', 'team2_def_yardsPerRushAttempt',
       'team1_off_yardsPerPass', 'team2_off_yardsPerPass',
       'team1_def_yardsPerPass', 'team2_def_yardsPerPass',
       'team1_off_rushingYards', 'team2_off_rushingYards',
       'team1_def_rushingYards', 'team2_def_rushingYards',
       'team1_off_netPassingYards', 'team2_off_netPassingYards',
       'team1_def_netPassingYards', 'team2_def_netPassingYards',
       'team1_off_firstDowns', 'team2_off_firstDowns', 'team1_def_firstDowns',
       'team2_def_firstDowns', 'team1_off_convertedthirdDowns',
       'team2_off_convertedthirdDowns', 'team1_def_convertedthirdDowns',
       'team2_def_convertedthirdDowns', 'team1_off_totalthirdDowns',
       'team2_off_totalthirdDowns', 'team1_def_totalthirdDowns',
       'team2_def_totalthirdDowns', 'team1_off_convertedfourthDowns',
       'team2_off_convertedfourthDowns', 'team1_def_convertedfourthDowns',
       'team2_def_convertedfourthDowns', 'team1_off_totalfourthDowns',
       'team2_off_totalfourthDowns', 'team1_def_totalfourthDowns',
       'team2_def_totalfourthDowns', 'team1_off_passCompletes',
       'team2_off_passCompletes', 'team1_def_passCompletes',
       'team2_def_passCompletes', 'team1_off_passAttempts',
       'team2_off_passAttempts', 'team1_def_passAttempts']
X_train = merged_df[(merged_df.year>2004) & (merged_df.year < 2020)][X_columns]
y_train = (merged_df[(merged_df.year>2004) & (merged_df.year < 2020)].team1_pts \
           > merged_df[(merged_df.year>2004) & (merged_df.year < 2020)].team2_pts)
X_test = merged_df[merged_df.year>=2020][X_columns]
y_test = (merged_df[merged_df.year>=2020].team1_pts > merged_df[merged_df.year>=2020].team2_pts)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

models = {
    "Baseline Model" : HigherELOWins(),
    "Logistic Regression" : LogisticRegression(max_iter=10000),
    "Decision Tree" : DecisionTreeClassifier(max_depth=100, max_leaf_nodes=6),
    "Gradient Boost" : GradientBoostingClassifier(loss='log_loss'),
    "Random Forest" : RandomForestClassifier(n_estimators=100, max_depth=100, max_leaf_nodes=6, random_state=1)
}

def probability_mse(estimator, X, y) :
    return mean_squared_error(estimator.predict_proba(X)[:,1], y.apply(int))

scores = dict()
for key, value in models.items() :
    scores[key] = cross_val_score(value, X_train, y_train, scoring=probability_mse).mean()
```

```python
fig, ax = plt.subplots()

ax.bar(scores.keys(), scores.values(), label=scores.keys())

ax.set_ylabel('MSE')
ax.set_title('Cross-Validation MSE Scores of Models 2002-2019')
plt.xticks(rotation='vertical')
plt.savefig('average_mse_cross_validation.png')
plt.show()
```

#### Fit data to full training set and predict 2020-2023 games.

```python
scores_df = pd.DataFrame(index = [2020, 2021, 2022, 2023], columns = models.keys())

for key, value in models.items() :
    value.fit(X_train, y_train)

for year in range(2020, 2024, 1) :
    year_df = merged_df[merged_df.year==year]
    X = year_df[X_columns]
    y = (year_df.team1_pts > year_df.team2_pts)
    for key, value in models.items() :
        scores_df.loc[year, key] = probability_mse(value, X, y)
```

```python
points = ['ko-', 'bv-', 'g^-', 'ms-', 'rx-']
for i in range(0, 5) :
    plt.plot(scores_df.iloc[:, i], points[i], label=scores_df.columns[i])
plt.xticks([2020, 2021, 2022, 2023])
plt.legend()
plt.xlabel("Year")
plt.ylabel("MSE")
plt.savefig("test_MSE_scores.png")
plt.show()
```
